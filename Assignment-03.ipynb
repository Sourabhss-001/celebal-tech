### Histogram:

```python
sns.histplot(data = df, x='Profit')
plt.xlabel('Profit USD')
plt.title('Profit Distribution', c='r')
```

```python
sns.histplot(df, x='Profit', kde=True, color = 'hotpink')
plt.xlabel('Profit USD')
plt.title('Profit Distribution', c='r', weight = 'bold')
```

```python
sns.histplot(data = df, 
             x = 'Profit', 
             hue = 'Year', 
             kde = True, 
             palette=sns.color_palette('colorblind'))

plt.title('Count of Profit')
```

### Barplot :

```python
sns.barplot(data=df, x='Country', y = 'Sales')
```

```python
# Set title
plt.title('', weight='bold')

# Make barplot of sale price by product
sns.barplot(data=df, x='Product', y='Units Sold', color = 'purple')
```

```python
sns.barplot(data = df,
            x = 'Product',
            y = 'Gross Sales',
            hue = 'Year',
            palette = 'dark:blue')
            
```

### Boxplot:

```python
sns.catplot(df, x = 'Product', y= 'Profit', kind = 'box', palette = 'winter' )
```

### Pie Plot

```python
# Setting up variables for pie chart
discount_band_counts = df['Discount Band'].value_counts().reset_index(name='Count')
discount_band_counts['Proportion'] = discount_band_counts['Count'] / discount_band_counts['Count'].sum() * 100

# Create the pie chart
plt.pie(discount_band_counts['Proportion'], labels=discount_band_counts['Discount Band'], autopct="%1.1f%%", shadow = True)
plt.title('Proportion of Discounts by Discount Band')
plt.show()
```

### Joint plot (scattered univariate with distributions) for sales and profit:

```python
# Create scatter plot
sns.scatterplot(data=df, x='Sales', y='Profit')

# Change labels for better readability
plt.ylabel('Profit (USD)')
plt.title('Sales versus Profit', weight='bold')
```

```python
sns.scatterplot(data=df, 
                x='Sales', 
                y='Profit', 
                hue='Discount Band',  
                style='Year',
                palette='colorblind')

# Change labels for better readability
plt.ylabel('Profit (USD)')
plt.title('Sales versus Profit', weight='bold')
```

```python
sns.scatterplot(data=df, 
                x='Sales', 
                y='Profit', 
                size='Discount Band',  
                hue='Year',
                palette='colorblind')

# Change labels for better readability
plt.ylabel('Profit (USD)')
plt.title('Sales versus Profit', weight='bold')
```

### Adding a regression line to the join plot:


```python
sns.jointplot(df, 
              x='Sales', 
              y='Profit', 
              kind='reg',
              color='green'
             )
```

### Violin plot showing the gross sales for each discount band:


```python
sns.violinplot(data=df, 
               x='Discount Band',
               y='Gross Sales',
               palette='colorblind')
```

```python
sns.violinplot(data=df, 
               x='Discount Band',
               y='Gross Sales',
               hue='Year',
               split=True,
               inner='quart',
               palette='colorblind')
```
