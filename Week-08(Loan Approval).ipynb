### 4. Encoding Categorical Variable

```python
df.drop("Loan_ID", axis=1, inplace=True)
```

```python
df.describe
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()

df['Gender'] = LE.fit_transform(df['Gender'])
df['Married'] = LE.fit_transform(df['Married'])
df['Education'] = LE.fit_transform(df['Education'])
df['Self_Employed'] = LE.fit_transform(df['Self_Employed'])
df['Property_Area'] = LE.fit_transform(df['Property_Area'])
df['Loan_Status'] = LE.fit_transform(df['Loan_Status'])
```

```python
df
```

```python
df['Dependents'].unique()
```

```python
df['Dependents'] = pd.to_numeric(df['Dependents'], errors='coerce').fillna(4).astype(int)
```

```python
df
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df['ApplicantIncome'])
```

```python
# remove Applicantincome greater than 10000
df = df[df['ApplicantIncome'] < 10000]
```

```python
import seaborn as sns
sns.boxplot(df['ApplicantIncome'])
```

```python
# separate depandent variable
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# transform into same scale
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x
```

```python
y
```

```python
#train test split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# create model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```

```python
# make prediction
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
# confusion matrix and accuracy
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
# heatmap of cm
import seaborn as sns
sns.heatmap(cm, annot=True)
```

```python
# actual v/s predictive data
new_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
new_df
```
